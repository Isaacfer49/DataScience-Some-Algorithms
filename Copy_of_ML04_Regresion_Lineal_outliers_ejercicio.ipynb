# Regresión lineal con mínimos cuadrados: valores atípicos

A menudo nuestros datos presentarán valores atípicos, generados por mediciones de sensores defectuosas o fluctuaciones aleatorias causadas por algún factor externo al sistema que estamos analizando. En este ejemplo, se generan datos artificialmente con un valor atípico.


**Objetivo:** Encontrar la línea que mejor ajuste a los datos con y sin el valor atípico, empleando la solución de mínimos cuadrados.

*   Discutir por qué los resultados son diferentes.
*   Describir qué se podría hacer para manejar los datos atípicos.



```python
import numpy as np
import matplotlib.pyplot as plt
import random
import copy

# asignar semilla
random.seed(200)
np.random.seed(200)

# Generamos datos artificiales
X_h = np.random.uniform(0,3,size=(10,1))
Y_h = 5 * (X_h) + np.random.normal(0, 1, size=(10,1))

# generamos valor atípico
Y_outlier = copy.copy(Y_h)  # copia de la variable dependiente
Y_outlier[6] = 50.0         # valor atípico


plt.title("Datos generados artificialmente con un valor atípico")
plt.scatter(X_h, Y_outlier, color="red")
plt.scatter(X_h, Y_h, color="blue")
plt.show()
```

```python
X_h = np.c_[np.ones((X_h.shape[0], 1)), X_h]

# Compute the least squares solution
theta_best = np.linalg.inv(X_h.T.dot(X_h)).dot(X_h.T).dot(Y_outlier)

# Print the calculated coefficients (intercept and slope)
print("Intercept (bias):", theta_best[0][0])
print("Slope:", theta_best[1][0])
```

```python
x_p = X_h[:,1]
```

```python
# Extract intercept and slope
intercept = theta_best[0][0]
slope = theta_best[1][0]

# Plot the data points
plt.scatter(x_p, Y_outlier, label="Data Points")

# Plot the linear regression line
y_line = intercept + slope * x_p
plt.plot(x_p, y_line, color='red', label="Linea de regression lineal")

# Add labels and legend
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regression lineal con minimos cuadrados utilizando el outlier ')
plt.legend()

# Show the plot
plt.show()
```

```python
#Minimos cuadrados sin el outlier
X_h = np.c_[np.ones((X_h.shape[0], 1)), X_h]


theta_best = np.linalg.inv(X_h.T.dot(X_h)).dot(X_h.T).dot(Y_h)

print("Intercept (bias):", theta_best[0][0])
print("Slope:", theta_best[1][0])
```

```python
x_p = X_h[:,1]
```

```python
#Extraer la interseccion y la pendiente
intercept = theta_best[0][0]
slope = theta_best[1][0]

#PLOTEAMOS
plt.scatter(x_p, Y_h, label="Data Points")

#PLOTEAMOS LA LINEA
y_line = intercept + slope * x_p
plt.plot(x_p, y_line, color='red', label="Linea de regression lineal")


plt.xlabel('X')
plt.ylabel('y')
plt.title('Regression lineal con minimos cuadrados SIN el outlier ')
plt.legend()


plt.show()
```
